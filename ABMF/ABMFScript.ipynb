#### Steps
- Import the csv file into a pandas df
- Clean the table name and remove extra symbols, spaces, capital letters
- write the create table SQL statement
- Import the data into the DB

```python
# importing libraries
import os
import pandas as pd
import numpy as np
import glob
import psycopg2
import sqlalchemy
import shutil
```

```python
# Grabbing all files from the main directory
pd.reset_option('^display.', silent=True)
os.chdir(r"C:\\Users\\Tobiloba\\Desktop\\ABMF\\Main")
# src_files = os.listdir(src)

globbed_files = glob.glob("*.csv") #creates a list of all csv files

globbed_files
```

```python
# Comparing backup files with main files to avoid duplication of files
for files in backup_files:
    if files in globbed_files:
        globbed_files.remove(files)

print(globbed_files)
```

```python
# Importing the file and deleting the file from directory after import
data = []
for csv in globbed_files:
    frame = pd.read_csv(csv, on_bad_lines='skip')
    frame['filename'] = os.path.basename(csv)
    data.append(frame)
    print("{} , File Data taken".format(csv))
    
    # sending file to a backup directory
    full_file_name = os.path.join("C:\\Users\\Tobiloba\\Desktop\\ABMF\\Main", csv)
    if os.path.isfile(full_file_name):
        shutil.copy(full_file_name, "C:\\Users\\Tobiloba\\Desktop\\ABMF\\Backup")
    print("{}, moved to backup".format(csv))
    
    # removing the file from the current directory
    os.remove(csv)    # might have to put this on the loop level to delete all files at the ending 
    print("{} , moved to database".format(csv))
    df = pd.concat(data)
```

```python
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        return conn
    except Error as e:
        print(e)

    return conn
```

```python
def create_table(conn, create_table_sql):
    """ create a table from the create_table_sql statement
    :param conn: Connection object
    :param create_table_sql: a CREATE TABLE statement
    :return:
    """
    try:
        c = conn.cursor()
        c.execute(create_table_sql)
    except Error as e:
        print(e)
```
