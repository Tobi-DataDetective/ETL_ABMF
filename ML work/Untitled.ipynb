```python
def none_working_time():
    none_working_hrs = pd.date_range('2022-09-01 18:00:00','2022-09-02 06:00:00', freq='1min')
    none_working_hrs.freq=None
    none_working_hrs

    none_working_hrs.astype(str)

    none_working_hrs_df = pd.DataFrame(none_working_hrs, columns=['Time'])

    none_working_hrs_df

    none_working_hrs_df['new_date'] = [d.date() for d in none_working_hrs_df['Time']]
    none_working_hrs_df['new_time'] = [d.time() for d in none_working_hrs_df['Time']]

    none_working_hrs_df.drop(columns=['Time', 'new_date'], inplace=True)
    none_working_hrs_df

    none_working_times = []
    for time in none_working_hrs_df.new_time:
        t = time.strftime("%H:%M:%S")
        none_working_times.append(t)
    return none_working_times

# none_working_time()
```

```python
def reformed_df(df):
    
    df.drop(columns=['filename'], axis=1, inplace=True)    

    df['Time']= pd.to_datetime(df['Time'])

    df['Time'] = df['Time'].dt.floor('T')

    df['Time'] = df['Time'].dt.floor('Min')

    df['new_date'] = [d.date() for d in df['Time']]
    df['new_time'] = [d.time() for d in df['Time']]
     
#     current_time_frame(df)
#     df['Time'].duplicated().any()
#     df['Time']= pd.to_datetime(df['Time'])
    df['Day_of_week']=df.Time.dt.strftime('%A')
    df = df[~df.Day_of_week.isin(['Saturday','Sunday'])]    
    df = df[~df.new_time.isin(none_working_time())]    
    return(df)
# df.drop_duplicates()
```

```python
def current_time_frame(df):
    
    df['Time']= pd.to_datetime(df['Time'])
    df['Time'] = df['Time'].dt.floor('T')
    df['Time'] = df['Time'].dt.floor('Min')
    df['new_date'] = [d.date() for d in df['Time']]
    df['new_time'] = [d.time() for d in df['Time']]
    # time = df['new_time'].strftime("%H:%M:%S")
    times = []
    for time in df.new_time:
        t = time.strftime("%H:%M:%S")
        times.append(t)
    return(times)
```
