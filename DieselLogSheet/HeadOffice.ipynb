```python
import os
import pandas as pd
import numpy as np
import glob
pd.reset_option('^display.', silent=True)
os.chdir(r"C:\\Users\\Tobiloba\\Desktop\\DieselLogSheet\\Head_office")

globbed_files = glob.glob("*.xlsx") #creates a list of all csv files

globbed_files
```

```python
data = []
for xlsx in globbed_files:
    frame = pd.read_excel(xlsx)
    frame['filename'] = os.path.basename(xlsx)
    data.append(frame)
    df = pd.concat(data)
    

site = df['filename']
```

```python
site = df['filename'].str.split('_', expand=True)[[0]]
```

```python
conditions = [(site[0] == 'HeadOffice')]

# create a list of the values we want to assign for each condition
values = ['Head Office']

# create a new column and use np.select to assign values to it using our lists as arguments
df['site'] = np.select(conditions, values)
# capitalizing the site names 
df['site'] = df['site'].str.capitalize()
```

```python
df = df.iloc[:, [0, 4,5,10,15]]
df.rename(columns = {'DATE':'Date', 'DIESEL TANK OPENING QUANTITY IN LITRES':'Diesel_Tank_Opening_Quantity', 'QUANTITY of DIESEL SUPPLIED (if diesel was supplied today)':'Quantity_of_Diesel_Supplied'\
                    , 'DIESEL TANK CLOSING QUANTITY IN LITRES':'Diesel_Tank_Closing_Quantity', 'DIFFERENCE IN DIESEL QUANTITY\n( USAGE)':'Difference_In_Diesel_Quantity'}, inplace = True)

df
```

```python
# removing all na values based on Diesel_Tank_Opening_Quantity
df = df[df['Diesel_Tank_Opening_Quantity'].notna()]
#drop rows that contain NEPA 
df = df[df.Diesel_Tank_Opening_Quantity != "NEPA"]
df = df[df.Date != "Date (DD/MM/YYYY)"]
# Replacing all NaN values with 0
df.fillna(0, inplace=True)
df.replace(['NIL'],[0], inplace=True)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
df = df.replace('LITRES','', regex=True)
df = df.replace('LTRES','', regex=True)
df = df.replace('was supplied','', regex=True)
df = df.replace('NO',0, regex=True)

df
```

```python
# Structuring the datatypes
df['Date'] = df['Date'].astype('datetime64')

df['Diesel_Tank_Opening_Quantity'] = df['Diesel_Tank_Opening_Quantity'].astype('int')

df['Quantity_of_Diesel_Supplied'] = df['Quantity_of_Diesel_Supplied'].astype('int')

df['Diesel_Tank_Closing_Quantity'] = df['Diesel_Tank_Closing_Quantity'].astype('int')

df['Difference_In_Diesel_Quantity'] = df['Difference_In_Diesel_Quantity'].astype('int')

df['Difference_In_Diesel_Quantity_Calc'] = df['Diesel_Tank_Opening_Quantity'] - df['Diesel_Tank_Closing_Quantity']
# Checking if correct
df['Truth'] = df['Difference_In_Diesel_Quantity'] == df['Difference_In_Diesel_Quantity_Calc']

df['Truth'].value_counts()

n_steps = 2
def rolling_diff(x):
    return x.iloc[-1] - x.iloc[0]

df['Refill_Check']=df.Diesel_Tank_Opening_Quantity.rolling(window=n_steps).apply(rolling_diff).fillna(-1)

df['Refilling_Check2'] = df['Refill_Check'] > 0
df
```
