```python
import os
import pandas as pd
import numpy as np
import glob
pd.reset_option('^display.', silent=True)
os.chdir(r"C:\\Users\\Tobiloba\\Desktop\\DieselLogSheet\\main")

globbed_files = glob.glob("*.xlsx") #creates a list of all csv files

globbed_files
```

```python
data = []
for xlsx in globbed_files:
    frame = pd.read_excel(xlsx)
    frame['filename'] = os.path.basename(xlsx)
    data.append(frame)
    df = pd.concat(data)
    

site = df['filename']
```

```python
site = df['filename'].str.split('_', expand=True)[[0]]
```

```python
conditions = [ (site[0] == 'Aba')\
             ,(site[0] == 'IkejaSG')\
             ,(site[0] == 'IkejaBG')\
             ,(site[0] == 'Onitsha')\
              ,(site[0] == 'Tradefair')\
              ,(site[0] == 'Apapa')\
              ,(site[0] == 'OkeArin')\
              ,(site[0] == 'Oshodi')\
              ,(site[0] == 'Onitsha2')\
              ,(site[0] == 'HeadOffice')\
             ]

# create a list of the values we want to assign for each condition
values = ['Aba','Ikeja SG','Ikeja BG','Onitsha','Tradefair','Apapa','OkeArin','Oshodi','Onitsha2','Head Office']

# create a new column and use np.select to assign values to it using our lists as arguments
df['site'] = np.select(conditions, values)
# capitalizing the site names 
df['site'] = df['site'].str.capitalize()
```

```python
# dropping the first 12 rows
df.drop(index=list(range(0,14)), inplace=True)
df = df[['Unnamed: 0','Unnamed: 7','Unnamed: 8','Unnamed: 14','Unnamed: 19','site']]
```
